# EJEMPLO 3.1 INTERFERENCIA DE ONDA
supongamos que dejamos caer una piedra en un estanque y las olas irradian desde el lugar donde cayó. podríamos crear una representación simple de la física con una onda sinusoidal, extendiéndose en un ciclo uniforme, para representar la altura de las ondas en algún momento posterior. si el centro del círculo está en $ x1, y1$, entonces la distancia r1 al centrarse desde un punto $x, y $ es:
\begin{equation}
  r_{1}=\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
\end{equation}

y la onda sinusoidal para la altura es:
\begin{equation}
  \xi_{1}(x,y)=\xi_{0}sinkr_{1}
\end{equation}
donde $\xi_{0}$ es la amplitud de las ondas y $k$ es el vector de onda, relacionado con la longitud de onda $\lambda$ por $k=\frac{2\pi}{\lambda}$
ahora supongamos que dejamos caer otra piedra en el estanque, creando otro conjunto circular de ondas con la misma longitud de onda y amplitud pero centradas en un punto diferente $x_{z}, y_{z}$:
\begin{equation}
  \xi_{1}(x,y)=\xi_{0}sinkr_{1}
\end{equation}
con
\begin{equation}
  r_{1}=\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
\end{equation}
luego, suponiendo que las olas se agreguen de manera lineal (lo cual es una suposición razonable para las ondas de agua, siempre que no sean demasiado grandes), la altura total de la superficie en un punto x, y es:
\begin{equation}
  \xi_{1}(x,y)=\xi_{0}sinkr_{1}+\xi_{0}sinkr_{2}
\end{equation}

supongamos que la longitud de onda de la es $\lambda=5cm$, la amplitud es de $ 1 cm$ y los centros de los círculos están separados por $20 cm$. Aquí hay un programa para hacer una imagen de la altura sobre una región cuadrada de 1m del estanque. para hacer la imagen, creamos una matriz de valores que representan la altura $\xi$ en una cuadrícula de puntos y luego usamos esa matriz para hacer un diagrama de densidad. En este ejemplo usamos una cuadrícula de $500 * 500$ puntos para cubrir el cuadrado de $1 m$, que significa que los puntos de la cuadrícula tienen una separación de $100/500 = 0.2 cm$

```
from math import sqrt,sin,pi
from numpy import empty
from pylab import imshow,gray,show
```

```
longituddeonda=5.0
k=2*pi/longituddeonda
xi0=1.0
separacion=20.0
lado=100.0
puntos=500
espaciado=lado/puntos
#calculando las posiciones de los circulos
x1=lado/2+separacion/2
y1=lado/2
x2=lado/2-separacion/2
y2=lado/2
#hacer una matriz para almacenar las alturas
xi= empty([puntos,puntos],float)
#calculando los valores de la matriz
for i in range(puntos):
    y=espaciado*i
    for j in range (puntos):
        x=espaciado*j
        r1=sqrt((x-x1)**2+(y-y1)**2)
        r2=sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j]=xi0*sin(k*r1)+xi0*sin(k*r2)
imshow(xi,origin="lower",extent=[0,lado,0,lado])
gray()
show()
```
